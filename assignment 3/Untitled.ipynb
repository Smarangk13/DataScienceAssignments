```python
import pandas as pd

file = 'BreadBasket_DMS.csv'
df = pd.read_csv(file)

foods = pd.DataFrame(df.groupby('Item'))
```

```python
str res = ''
for i in foods[0]:
    print(i)
    x = input("type : ")
    res += str(i)+','
    if x == 'f':
        res += 'food\n'
    elif x == 'd':
        res += 'drink\n'
    else:
        res += 'unkown\n'
        
```
